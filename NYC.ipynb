```python
sales_manhattan = sales[sales['BOROUGH NAME'] == 'Manhattan']
```

```python
sales_manhattan.describe()
```

```python
# There were sold 18306 apartments in the Manhattan area.
```

### What was the average selling price?

```python
sales_manhattan['SALE PRICE'].mean()
```

```python
# The average selling price was 2.632.835$
```

### In which neighborhood in Manhattan was sold the most expensive apartment?

```python
sales_manhattan.sort_values(['SALE PRICE'], ascending = False).head(1)
```

```python
# Midtown CBD was the most expensive apartment sold at 22.1000.000$
```

### Which month of the year has the most sales? Plot the number of sales each month in a barplot.


```python
sales_manhattan.groupby('MONTH').count().sort_values(['SALE PRICE'], ascending = False).head(1)
```

```python
# The month with thw most sales was June.
```

```python
sales_manhattan.groupby('MONTH').count()['SALE PRICE'].plot.bar(figsize=(12, 6), color = 'magenta')
                                                    
```

### How many apartments were sold from parents to children? Which month of the year and in which borough were sold the most?

```python
sales_parents = sales[sales['SALE PRICE'] == 0] 
sales_parents.shape
```

```python
sales_parents.groupby('MONTH').count().sort_values(['SALE PRICE'], ascending = False).head(1)
```

```python
sales_parents.groupby('BOROUGH NAME').count().sort_values(['NEIGHBORHOOD'], ascending = False)
```

```python
sales_parents.groupby('BOROUGH NAME').count().sort_values(['NEIGHBORHOOD'], ascending = False)['NEIGHBORHOOD'].plot.bar(figsize=(12, 6), color = 'orange')
```
